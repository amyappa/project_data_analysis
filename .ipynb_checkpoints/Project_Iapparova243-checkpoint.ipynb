```python
country_noinfo=data1[data1['Inflation'].isna()==True]
country_noinfo2007=country_noinfo[(country_noinfo['date'] > '2007-01-01') & (country_noinfo['date'] < '2008-01-01')]
countries2007=country_noinfo2007['country'].unique()
#countries without information about inflation in 2007
print(*countries2007)
known_count=set(all_countries)-set(countries2007)
proportion_unknowncount=(len(countries2007)/len(all_countries))*100
print(proportion_unknowncount,'%')
```

### As we see 76% of countries in data do not have information about inflation in 2007. Since there is data for the period from 2007 to 2023, I think it is possible to start the analysis from 2008. Therefore, I will delete the data from all countries for 2007.

```python
indices_to_drop = data1[data1['date'].dt.year == 2007].index
data1.drop(indices_to_drop, inplace=True)
remaining_rows = data1[data1['date'].dt.year == 2007] #check to ensure that we dropped them

print(len(remaining_rows))
data1
```

```python
data1.isna().sum()
```

### As we can see, there are proportionally fewer NaN values left, but they remain. Since I will be researching inflation and its changes, filling in averages can distort possible patterns and trends and using interpolation cannot work as it assumes a linear relationship between existing neighbors, but in my data there are long missing sequences.
### So I will use the bfill (fills missing values with the next valid value) and ffill (fills remaining missing values with the previous valid value) methods and apply them separately to each country so as not to mix up the unique trends of the countries.

```python
original_data=data1.copy()
columns_to_fill = ['Open', 'High', 'Low', 'Close', 'Inflation']
for col in columns_to_fill:
    data1[col] = data1.groupby('country')[col].apply( #groups the data by country and for each column,
#it processes the data separately for each country
        lambda group: group.bfill().ffill() 
    ).reset_index(level=0, drop=True)
    
data1.isnull().sum()
```

# 2nd Hypothesis: 
## Countries with higher food inflation volatility experience more frequent supply shocks


I predict that countries where food prices have more frequent variability undergo higher inflation surges.

### Firstly, I transform 'average_annualized_food_volatility' and maximum_food_drawdown' into suitible format type

```python
data2['average_annualized_food_volatility'] = (data2['average_annualized_food_volatility']
    .str.replace('%', '', regex=False)  
    .str.replace('.', '')  
    .astype(float)  
    / 100
)
data2['maximum_food_drawdown'] = (data2['maximum_food_drawdown']
    .str.replace('%', '', regex=False) 
    .str.replace('.', '')  
    .astype(float)  
    / 100  
)
data2.head(7)
```

```python
data1_yearly = yearly_inflation.groupby(['country', 'ISO3', 'year']).agg({'Inflation': 'mean'}).reset_index()
merged_data = pd.merge(data1_yearly,data2, on=['country', 'year'], how='outer')
```

### I want to count the number of supply shocks by finding extreme cases where inflation deviated greatly from the average

```python
def detect_supply_shocks(group):
    mean_inflation = group.mean()
    std_inflation = group.std()
    shocks = (abs(group - mean_inflation) >  2*std_inflation).sum()
    return shocks

supply_shocks = (
    merged_data[['country', 'Inflation']]  
    .groupby('country')['Inflation']  
    .apply(detect_supply_shocks) 
    .reset_index(name='shock_count')  
)

print(supply_shocks)
```

```python
inflation_variability = merged_data.groupby('country')['Inflation'].std().reset_index(name='inflation_std_dev')
country_metrics = pd.merge(supply_shocks, inflation_variability, on='country')
country_metrics = pd.merge(country_metrics, data2[['country', 'average_annualized_food_volatility', 'maximum_food_drawdown']], on='country')
country_metrics
```

```python
# Correlation between volatility and supply shocks
correlation, p_value = pearsonr(country_metrics['average_annualized_food_volatility'], country_metrics['shock_count'])
print(f"Correlation between volatility and supply shocks: {correlation.round(3)} (p-value: {p_value.round(3)})")

# Linear regression model
x = country_metrics[['average_annualized_food_volatility', 'maximum_food_drawdown']]
y = country_metrics['shock_count']

model = LinearRegression()
model.fit(x, y)

print(f"Regression Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

### The findings reveal that price changes in food markets are not effective predictors of supply disruptions.
* Correlation: changes in food volatility do not appear to predict the number of supply shocks because correlation coefficient is too small
* P-value: p-value above 0.05 means the correlation is not statistically significant.
* Coefficient volatility: A 1-unit increase in volatility is associated with an increase of only 0.0046 supply shocks. This is a very small and likely insignificant effect.
* Intercept: When both volatility and drawdown are 0, the model predicts 0.724 supply shocks. This serves as the baseline value.

```python
# Scatterplot: Volatility vs. Supply shock count
fig_scatter = px.scatter(
    country_metrics,
    x='average_annualized_food_volatility',
    y='shock_count',
    title='Volatility vs. Supply shock count',
    labels={
        'average_annualized_food_volatility': 'Average annualized food volatility',
        'shock_count': 'Supply shock count'
    },
    hover_name='country',
    template='plotly_white'
)
fig_scatter.update_traces(marker=dict(size=10, color='blue'))
fig_scatter.show()

# Boxplot: Inflation variability by volatility group
merged_data['volatility_group'] = pd.qcut(
    merged_data['average_annualized_food_volatility'], 
    4,  # I make quartiles
    labels=['Low', 'Medium-Low', 'Medium-High', 'High']
)

boxplot_data = (
    merged_data.groupby('volatility_group', observed=True)['Inflation']
    .std() 
    .reset_index(name='Inflation variability')
)

fig_boxplot = px.bar(
    boxplot_data,
    x='volatility_group',
    y='Inflation variability',
    title='Inflation variability by volatility group',
    labels={'volatility_group': 'Volatility group', 'Inflation variability': 'Inflation variability (Std Dev)'},
    template='plotly_white'
)

fig_boxplot.show()
```

### Chart 1 
There isn’t a clear pattern. Both low and high volatility countries have similar numbers of supply shocks.

### Chart 2
* Countries with more stable food prices ("Low" volatility) have smaller changes in inflation.
* Countries with unstable food prices ("High" volatility) experience much bigger swings in inflation.

#### It means that when food prices are unpredictable, inflation also becomes unpredictable.

## Conclusion 

### - Supply shocks are not directly caused by price instability
